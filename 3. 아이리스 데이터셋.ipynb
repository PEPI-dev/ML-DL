- [사이킷런 데이터셋 페이지](https://scikit-learn.org/stable/modules/classes.html?highlight=datasets#module-sklearn.datasets)




```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

- sepal length in cm : 꽃 받침의 길이
- sepal width in cm : 꽃 받침의 너비
- petal length in cm : 꽃잎의 길이
- petal width in cm : 꽃잎의 너비

```python
data = iris['data']
data
```

```python
feature_names= iris['feature_names']
feature_names
```

```python
import pandas as pd
```

```python
df_iris = pd.DataFrame(data, columns=feature_names)
df_iris.head()
```

```python
target = iris['target']
df_iris['target'] = target
df_iris.head()
```

```python
df_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split(독립변수, 종속변수, 테스트사이즈, 시드값...)
X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target', 1),df_iris['target'],test_size=0.2,random_state=2023)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predic(X_test)
```

```python
print('정답률' ,accuracy_score(y_test,y_pred))
```

```python
y_pred = svc.predict([[6.4, 2.6, 5.1, 1.9]])
y_pred
```

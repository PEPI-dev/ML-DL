```python
# 문제.
# Type_of_Loan의 모든 대출 상품을 변수에 모두 저장
# 각 대출상품에 대한 이름으로 컬럼으 만들고 대충 상품이 있는 경우 해당 컬럼에 1을 저장(원핫 인코딩)
credit_df['Type_of_Loan'] = credit_df['Type_of_Loan'].str.replace('and ','')
```

```python
type_list = set(credit_df['Type_of_Loan'].str.split(', ').sum())
type_list
```

```python
for i in type_list:
  credit_df[i] = credit_df['Type_of_Loan'].apply(lambda x: 1 if i in x else 0)
```

# 2. lightGBM(LGBM)
- Microsoft에서 개발한 Gradient Boosting Framework
- 리프 중심 히스토리그램 기반 알고리즘
- GBM(Gradient Boosting Model) : 모델1을 통해 y 예측하고, 모델 2에 데이터를 넣ㅇ서 y예측하고 , 모델3에 넣어서 y 예측
- 작은 데이터셋에서도 높은 성능을 보이며, 특히 대용량 데이터셋에서 다른 부스팅알고리즘 보다 빠름
- 메모리 사용랴잉 상대적으로 적ㅇㄴ편
- 작은 데이터터셋을 사용할 경우 과적합 가능성 매우 큼(일반적으로 데이터가 최소 10000개 이상 사용해야 안정적)
- 조기 중단(early stopping) 지원

### 2-1. 리프 중심 히스토그램 기반 알고리즘
- 트리를 균형적으로 불한하는게 아니라 최대한 불균형하게 분할
-

```python
from lightgbm import LGBMClassifier
```

```python
base_model = LGBMClassifier(random_state=2023)
```

```python
base_model.fit(X_train,X_test)
```

```python
pred = base_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
proba = base_model.predict_proba(X_test)
proba
```

```python
roc_auc_score(y_test,proba,multi_calss='ovr')
```

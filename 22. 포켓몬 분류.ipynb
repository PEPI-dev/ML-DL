```python
# 필요한 모듈 준비
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
```

```python
# GPU 사용 확인
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
# 이미지 증강기법 사용
# data_transforms
# train / validation

data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomAffine(0, shear=10, scale=(0.8, 1.2)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor()
    ]),
    'validation': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor()
    ])
}
```

```python
# 데이터셋 객체 생성
# image_datasets
# train / validation

# # train 및 validation 클래스 디렉토리 목록 가져오기
# train_class_dirs = [d for d in os.listdir('train') if os.path.isdir(os.path.join('train', d))]
# validation_class_dirs = [d for d in os.listdir('validation') if os.path.isdir(os.path.join('validation', d))]

# # train 데이터셋 객체 생성
# image_datasets['train'] = datasets.ImageFolder('train', data_transforms['train'], classes =train_class_dirs)

# # validation 데이터셋 객체 생성
# image_datasets['validation'] = datasets.ImageFolder('validation', data_transforms['validation'], classes=validation_class_dirs)


image_datasets = {
    'train': datasets.ImageFolder('train', data_transforms['train']),
    'validation': datasets.ImageFolder('validation', data_transforms['validation'])
}
```

```python
# 데이터로더 생성
# dataloaders
# batch_size = 32
# shuffle =True

dataloaders = {
    'train': DataLoader(
        image_datasets['train'],
        batch_size=32,
        shuffle=True
    ),
    'validation': DataLoader(
        image_datasets['validation'],
        batch_size=32,
        shuffle=False
    )
}
```

```python
from torchvision.models import efficientnet_b4, EfficientNet_B4_Weights
from torchvision.models._api import WeightsEnum
from torch.hub import load_state_dict_from_url
```

```python
def get_state_dict(self, *args, **kwargs):
    kwargs.pop("check_hash")
    return load_state_dict_from_url(self.url, *args, **kwargs)
WeightsEnum.get_state_dict = get_state_dict
```

```python
model = models.efficientnet_b4().to(device)

# FC Layer 수정
# print(model)
for param in model.parameters():
    param.requires_grad = False # 가져온 파라미터 (W, b)를 업데이트하지 않음
# 149

model.classifier = nn.Sequential(
    nn.Linear(1792, 512),
    nn.ReLU(),
    nn.Linear(512, 149)
)

model = model.to(device)
model1 = model1.to(device)
# print(model)
```

```python
# 학습
# optimizer: Adam
# epochs: 10
optimizer = optim.Adam(model.classifier.parameters(), lr=0.001)
# optimizer = optim.Adam(model.parameters(), lr=0.001) # fc 파라미터와 다름

epochs = 10

for epoch in range(epochs+1):
    for phase in ['train', 'validation']:
        if phase == 'train':
            model.train()
        else:
            model.eval()

        sum_loss = 0
        sum_acc = 0

        for x_batch, y_batch in dataloaders[phase]:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)

            y_pred = model(x_batch)
            loss = nn.CrossEntropyLoss()(y_pred, y_batch)

            if phase == 'train':
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            sum_loss = sum_loss + loss.item()

            y_prob = nn.Softmax(1)(y_pred)

            y_pred_index = torch.argmax(y_prob, axis=1)

            accuracy = (y_batch == y_pred_index).float().sum() / len(y_batch) * 100
            sum_acc = sum_acc + accuracy

        avg_loss = sum_loss / len(dataloaders[phase])
        avg_acc = sum_acc / len(dataloaders[phase])

        print(f'{phase:10s}: Epoch {epoch:4d}/{epochs} Loss: {avg_loss:.6f} Accuracy: {avg_acc:.2f}%')
```

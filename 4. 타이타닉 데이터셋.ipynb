### 4-3. 라벨인코딩(Label Encoding)
문자데이터를 수치로 변환

```python
df['Sex'].value_counts()
```

```python
# 남자는 1, 여자는 0으로 라벨 인코딩

# df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})
# label_counts = df['Sex'].value_counts()
# label_counts

def convert_sex(data):
  if data == 'male':
    return 1

  elif data == 'female':
    return 0

```

```python
df['Sex'] = df['Sex'].apply(convert_sex)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Embarked'].value_counts()
```

```python
embarked = le.fit_transform(df['Embarked'])
embarked
```

### 4-4. 원 핫 인코딩(One Hot Encoding)
- 독립적인 데이터는 별도의 컬럼으로 분리하고 각각 컬럼에 해당 값에만 1, 나머지는 0값으로 갖게 하는 방법


```python
df['Embarked_num'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
pd.get_dummies(df['Embarked_num'])
```

```python
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
df.head()
```
